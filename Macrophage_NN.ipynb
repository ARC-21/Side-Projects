```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# main_marker_data = pd.read_csv('[MainMarkers]_0-25.txt')
basic_o2_data = pd.read_csv('[MainMarkers]_vs_[O2_0]_0-25.txt', sep='\t')
basic_o2_data.drop(columns=['[IL4]_0'], inplace=True)

# display(main_marker_data.head())
display(basic_o2_data.head())
print(basic_o2_data.shape)
```

```python
initial = basic_o2_data[basic_o2_data['Time'] == 0]
initial.drop(columns=['Time'], inplace=True)
display(initial.head())
print(f'initial.shape: {initial.shape}')

final = basic_o2_data[basic_o2_data['Time'] == 20000]
final.drop(columns=['Time'], inplace=True)
display(final.head())
print(f'y.shape: {final.shape}')
```

```python
m1 = final[['[IFNG]', '[iNOS]', '[TNFa]', '[CXCL9]', '[IL12]', '[mCXCL10]']].sum(axis=1)
m2 = final[['[VEGF]', '[ARG1]', '[IL10]']].sum(axis=1)

m1_m2_ratios = m1 / m2# pd.concat([m1, m2], axis=1)

display(m1_m2_ratios.head()) # il12 and ifng
print(m1_m2_ratios.shape)

# plt.plot(basic_o2_data['[O2]_0'][basic_o2_data['Time'] == 10000], m1_m2_ratios)
# plt.title('M1/M2 Ratios vs O2 Initial Concentrations')
# plt.xlabel('O2 Initial Concentration')
# plt.ylabel('M1/M2 Ratio')
```

```python
from sklearn.preprocessing import StandardScaler
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

scaler = StandardScaler()
initial = scaler.fit_transform(initial)
```

```python
tensor_X = torch.tensor(initial, dtype=torch.float32)
tensor_Y = torch.tensor(m1_m2_ratios.to_numpy(), dtype=torch.float32)
```

```python
from torch.utils.data import Dataset

class MacrophageData(Dataset):
    def __init__(self, X, Y):
        self.X = X
        self.Y = Y

    def __len__(self):
        return len(self.Y)

    def __getitem__(self, idx):
        return self.X[idx], self.Y[idx]
```

```python
custom_data = MacrophageData(tensor_X, tensor_Y)

train_dataloader = DataLoader(custom_data, batch_size=16, shuffle=True)
test_dataloader = DataLoader(custom_data, batch_size=16, shuffle=False)
```

```python
for X, y in train_dataloader:
    print(f"X: {X.shape}\nY: {y.shape}")
    break
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

def reset_weights(model):
    for layer in model.children():
        if hasattr(layer, 'reset_parameters'):
            layer.reset_parameters()

class RegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_stack = nn.Sequential(
            nn.Linear(10, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.linear_stack(x).squeeze(-1)

model = RegressionModel()
reset_weights(model)

loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
```

```python
def train(dataloader, model, loss_fn, optimizer):
    model.train()
    total_loss = 0
    for batch, (X, y) in enumerate(dataloader):
        pred = model(X)
        loss = loss_fn(pred, y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        if batch % 10 == 0:
            print(f"\tBatch {batch}, Loss: {loss.item():.6f}")

    return total_loss / len(dataloader)
```

```python
from sklearn.metrics import r2_score

def test(dataloader, model, loss_fn):
    model.eval()
    test_loss = 0
    num_batches = len(dataloader)

    all_preds = []
    all_targets = []

    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()

            all_preds.append(pred.cpu().numpy())
            all_targets.append(y.cpu().numpy())

    test_loss /= num_batches

    all_preds = np.concatenate(all_preds, axis=0)
    all_targets = np.concatenate(all_targets, axis=0)

    r2 = r2_score(all_targets, all_preds)

    print(f"\t\tAvg loss: {test_loss:.6f} | R^2 Score: {r2:.6f}")

    return test_loss, r2
```

```python
import numpy as np
import matplotlib.pyplot as plt

epochs = 200
train_losses = []
test_losses = []
r2_scores = []

for t in range(epochs):
    print(f"Epoch {t+1} ------------------------------")
    train_loss = train(train_dataloader, model, loss_fn, optimizer)
    test_loss, r2 = test(test_dataloader, model, loss_fn)

    train_losses.append(train_loss)
    test_losses.append(test_loss)
    r2_scores.append(r2)

    print(f"\tTraining Loss: {train_loss:.6f} | Validation Loss: {test_loss:.6f} | R^2 Score: {r2:.6f}\n")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(1, epochs+1), train_losses, label='Training Loss')
plt.plot(range(1, epochs+1), test_losses, label='Validation Loss', linestyle='dashed')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, epochs+1), r2_scores, label='R^2 Score', color='green')
plt.xlabel('Epoch')
plt.ylabel('R^2 Score')
plt.title('R^2 Score')
plt.legend()

plt.show()
```
